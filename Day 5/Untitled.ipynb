```python
for building in range(1, 4):      # Outer Loop
    for floor in range(1, 5):     # Middle Loop
        for room in range(1, 6):  # Inner Loop
            print("Building", building , "Floor" , floor , "Room" , room)
```

```python


buildings = int(input("Enter number of buildings: "))
floors = int(input("Enter number of floors in each building: "))
rooms = int(input("Enter number of rooms on each floor: "))

for building in range(1, buildings + 1):
    print("\nBuilding", building)

    for floor in range(1, floors + 1):
        print("  Floor", floor)

        for room in range(1, rooms + 1):
            print("    Deliver medicine to Room", room)
```
