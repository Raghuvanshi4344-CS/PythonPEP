```python
# Abstaction : Create An abstract class Vehicle with an abstract methid fule_type(). Create PetrolCar and ElectricCar subclasses, each implementing fule_type() differently
from abc import ABC, abstractmethod
class Vehicle(ABC):

    @abstractmethod
    def fuel_type(self):
        pass


class PetrolCar(Vehicle):
    def fuel_type(self):
        print("Fuel Type: Petrol")


class ElectricCar(Vehicle):
    def fuel_type(self):
        print("Fuel Type: Electricity")


p = PetrolCar()
e = ElectricCar()

p.fuel_type()
e.fuel_type()
```

```python
#Inheritance - Create a base class Person with name and age. Create Teacher and Student subclasses,each adding one extra attribute(e.g.  subject For Teacher, roll_ni for Student ) and overriding a common method role() to describe who they are.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def role(self):
        print("I am a Person")

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def role(self):
        print("I am a Teacher")


class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age)
        self.roll_no = roll_no

    def role(self):
        print("I am a Student")


t = Teacher("Rahul", 35, "Math")
s = Student("Bibek", 20, 101)

t.role()
print(t.name, t.age, t.subject)

s.role()
print(s.name, s.age, s.roll_no)
```

```python
#Polymorphism - Create Square and Triangle classes, each with a draw() method that prints a different simple pattern (e.g. using *). Put objects of both in a list and loop through ,calling draw() on each.
class Square:
    def draw(self):
        print("***")
        print("***")
        print("***")


class Triangle:
    def draw(self):
        print("*")
        print("**")
        print("***")


shapes = [Square(), Triangle()]

for shape in shapes:
    shape.draw()
    print()
```

```python
# Smart Library System
# Design a system to issue and return books. Store book availability safely , allow members to borrow/ return books ,and calculate late fees if applicable
class Book:
    def __init__(self, title):
        self.title = title
        self.__available = True

    def issue(self):
        if self.__available:
            self.__available = False
            print("Book Issued")
        else:
            print("Book Not Available")

    def return_book(self):
        self.__available = True
        print("Book Returned")


class Member:
    def borrow(self, book):
        book.issue()

    def return_book(self, book, days):
        book.return_book()
        if days > 7:
            print("Late Fee =", (days - 7) * 10)


b = Book("Python")
m = Member()

m.borrow(b)
m.borrow(b)
m.return_book(b, 10)
```

```python
#Airline Ticket Reservation
#Create Economy ,Business and first Class tickrts. Calculate ticket fare based on class and baggage weight.Print boarding details.
class Ticket:
    def __init__(self, name, baggage):
        self.name = name
        self.baggage = baggage

class Economy(Ticket):
    def fare(self):
        print("Fare:", 5000 + self.baggage * 50)

class Business(Ticket):
    def fare(self):
        print("Fare:", 10000 + self.baggage * 50)

class FirstClass(Ticket):
    def fare(self):
        print("Fare:", 15000 + self.baggage * 50)

e = Economy("Bibek", 10)
b = Business("Govind", 15)
f = FirstClass("Hitesh", 20)

for ticket in [e, b, f]:
    print("Passenger:", ticket.name)
    ticket.fare()
    print("Boarding Successful\n")
```

```python
#Restaurant Billing 
#Customers order food items. Generate the bill ,apply gst and discount and support Dine-in and takeaway orders differently.
class Order:
    def __init__(self, amount):
        self.amount = amount

class DineIn(Order):
    def bill(self):
        total = self.amount + self.amount*0.05 - 50
        print("Dine-In Bill:", total)

class TakeAway(Order):
    def bill(self):
        total = self.amount + self.amount*0.05
        print("TakeAway Bill:", total)

d = DineIn(1000)
t = TakeAway(1000)

d.bill()
t.bill()
```
