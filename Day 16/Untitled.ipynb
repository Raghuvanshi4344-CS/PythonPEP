```python
#Create a Student class with roll_no,name,marks.Add a method get_greade()that returns "A" if marks >= 90 , "b" if >=75 , else "c"
class Student:
    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks

    def get_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        else:
            return "C"
s = Student(101, "Bibek", 99)
print("Roll No:", s.roll_no)
print("Name:", s.name)
print("Marks:", s.marks)
print("Grade:", s.get_grade())
```

```python
# create a Rectangle class with length and width .Add methods area() and perimeter().
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print("Area =", self.length * self.width)

    def perimeter(self):
        print("Perimeter =", 2 * (self.length + self.width))

r = Rectangle(10, 5)

r.area()
r.perimeter()
```

```python
#Q6- (Challenge - ties to earlier collection topic ).Create a Library class that holds a list of Book objects(from Q1).Add methods add_book and show_all_books()

class Book:
    def __init__(self, title):
        self.title = title


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def show_all_books(self):
        for book in self.books:
            print(book.title)


b1 = Book("Python")
b2 = Book("Java")
b3 = Book("C++")

lib = Library()

lib.add_book(b1)
lib.add_book(b2)
lib.add_book(b3)

lib.show_all_books()
```
