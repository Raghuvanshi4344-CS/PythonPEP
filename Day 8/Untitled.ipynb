```python
# Global variable
city = "Bhubaneswar"

# Function
def display_city():
    print("Inside function:", city)

# Function call
display_city()

# Outside the function
print("Outside function:", city)
```

```python
# Global variable
balance = 1000

def deposit():
    global balance
    balance = balance + 500
    print("Balance after deposit:", balance)

def withdraw():
    global balance
    balance = balance - 200
    print("Balance after withdraw:", balance)

print("Initial Balance:", balance)

deposit()
withdraw()

print("Final Balance:", balance)
```

```python
#prime number
prime = lambda n: all(n % i != 0 for i in range(2, n)) if n > 1 else False

numbers = [2, 3, 4, 5, 6, 7, 8, 9, 10]

result = list(filter(prime, numbers))

print(result)
```
