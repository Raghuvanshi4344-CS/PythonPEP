```python
def sum_n(n):
    if n == 0:        # Base case
        return 0
    return n + sum_n(n - 1)   # Recursive call

num = int(input("Enter a number: "))

if num < 0:
    print("Please enter a positive number.")
else:
    print("Sum =", sum_n(num))
```

```python
#recursive list sum
#write a recursive function to find sum of all numbers in a list without using sum() or lops.

def list_sum(lst):
    if len(lst) == 0:      # Base case
        return 0
    return lst[0] + list_sum(lst[1:])

numbers = [10, 20, 30, 40, 50]

print("List:", numbers)
print("Sum =", list_sum(numbers))
```

```python
#lambdaa+filter()
#given student (Name,marks,age) display only students with marks>80 and age<22 uisng lambda and filer()

students = [
    ("Rahul", 85, 20),
    ("Priya", 78, 21),
    ("Amit", 92, 23),
    ("Neha", 88, 19)
]
result = list(filter(lambda s: s[1] > 80 and s[2] < 22, students))
for student in result:
    print(student[0])
```
