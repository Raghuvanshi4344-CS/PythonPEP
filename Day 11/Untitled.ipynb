```python
#find the frequency of each character in a string 
s = input("Enter a string: ")

for ch in set(s):
    print(ch, ":", s.count(ch))
```

```python
#convert a snake_case string to Camelcase.
s = input("Enter snake_case string: ")

words = s.split("_")

camel = ""

for word in words:
    camel = camel + word.capitalize()
print("CamelCase:", camel)
```

```python
#check whether two strings are anagrams
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Not Anagram")
```

```python
#find the longest word in a sentence
sentence = input("Enter a sentence: ")

words = sentence.split()


longest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
```

```python
#q1-netflix content dashboard
#concept -indexing and slicing
#shows list provided. Display first , last , top3,last 3,alternate shows,reverse using slicing 
# Netflix Content Dashboard

shows = ["GOT", "Wednesday", "Money Heist", "Squid Game", "Dark", "Lucifer"]

print("First Show:", shows[0])
print("Last Show:", shows[-1])

print("Top 3 Shows:", shows[:3])
print("Last 3 Shows:", shows[-3:])

print("Alternate Shows:", shows[::2])

print("Reverse Shows:", shows[::-1])
```

```python
#q2 hospital patient monitoring
# conceot-traversal
#Heart rates list.Print patients with heart rate >100bpm and count them.
heart_rates = [72, 105, 98, 120, 88, 110, 95]

count = 0
for rate in heart_rates:
    if rate > 100:
        print(rate, "bpm")
        count = count + 1
print("Total Patients:", count)
```
