```python
#Method	Purpose
#add(x)	Add one element
#update(iterable)	Add multiple elements
#remove(x)	Remove element (error if absent)
#discard(x)	Remove element (no error if absent)
#pop()	Remove a random element
#clear()	Remove all elements
#copy()	Copy the set

# Set Methods in One Program

s = {10, 20, 30}

print("Original Set:", s)

# 1. add() - Adds one element
s.add(40)
print("After add(40):", s)

# 2. update() - Adds multiple elements
s.update([50, 60])
print("After update([50, 60]):", s)

# 3. remove() - Removes an element
s.remove(20)
print("After remove(20):", s)

# 4. discard() - Removes an element (no error if not found)
s.discard(100)      # 100 is not present, so no error
print("After discard(100):", s)

# 5. pop() - Removes a random element
removed = s.pop()
print("Removed by pop():", removed)
print("After pop():", s)

# 6. copy() - Creates a copy of the set
copy_set = s.copy()
print("Copied Set:", copy_set)

# 7. clear() - Removes all elements
s.clear()
print("After clear():", s)
```

```python
#q1 - Store the last 15 temperatures recorded during a day and display the highest ,lowest and average tempreture
temperatures = []

for i in range(15):
    temp = int(input("Enter temperature: "))
    temperatures.append(temp)

print("Highest Temperature:", max(temperatures))
print("Lowest Temperature:", min(temperatures))
print("Average Temperature:", sum(temperatures) / 15)
```
