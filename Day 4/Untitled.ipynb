```python
#print table using while loop
num = int(input("Enter a number: "))

i = 1

while i <= 10:
    print(num, "*", i, "=", num * i)
    i = i + 1
```

```python
#count digit of given input number
num = int(input("Enter a number: "))

count = 0

while num > 0:
    count = count + 1
    num = num // 10

print("Total digits =", count)
```

```python
#reverse a given numer using while loop

num = int(input("Enter a number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reverse =", reverse)
```

```python
#Check Palindrome using while loop
num = int(input("Enter a number: "))

temp = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if temp == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#sum of all digits

num = int(input("Enter a number: "))

sum = 0

while num > 0:
    digit = num % 10
    sum = sum + digit
    num = num // 10

print("Sum =", sum)
```

```python
#keep Taking Numbers Until User Enters 0

sum = 0

num = int(input("Enter a number (0 to stop): "))

while num != 0:
    sum = sum + num
    num = int(input("Enter a number (0 to stop): "))

print("Total Sum =", sum)
```

```python
#Print numbers from 1 to 20. Skip all multiples of 3

i = 1

while i <= 20:

    if i % 3 == 0:
        i = i + 1
        continue

    print(i)
    i = i + 1
```

```python
#Print numbers from 1 onwards.Skip multiples of 4.Stop the loop when the number becomes greater than 20.

i = 1

while True:

    if i > 20:
        break

    if i % 4 == 0:
        i = i + 1
        continue

    print(i)
    i = i + 1
```
