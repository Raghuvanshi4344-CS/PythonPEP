```python
#part E-Mini project 
#Library management System.Store books ,add books,search ,issue,return ,create backup, sort books ,and display final library using appropriate list methods.

library = ["Python", "Java", "C++", "DBMS","Os"]
print("Initial Library:", library)

library.append("Data Science")
print("\nAfter Adding Book:", library)

book = input("\nEnter book to search: ")
if book in library:
    print(book, "is available at position", library.index(book))
else:
    print(book, "is not available.")

issue = input("\nEnter book to issue: ")
if issue in library:
    library.remove(issue)
    print(issue, "issued successfully.")
else:
    print("Book not found.")

return_book = input("\nEnter book to return: ")
library.append(return_book)
print(return_book, "returned successfully.")

backup = library.copy()
print("\nLibrary Backup:", backup)

library.sort()
print("\nFinal Library:", library)

print("Total Books:", len(library))















```
