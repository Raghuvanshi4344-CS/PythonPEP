```python
num1=int(input("Enter first num: "))
opr=input("Enter Your operators(+,-,*,/,%): ")
num2=int(input("Enter second num: "))

if opr == "+":
   print(num1 + num2)
   
elif opr == "-":
   print(num1 - num2)

elif opr == "*":
   print(num1  * num2)

elif opr == "/":
   print(num1  / num2)

elif opr == "%":
   print(num1 % num2)

else:
    ("invalid input")
         
         
```
