```python
with open("intro.txt", "w") as file:
    file.write("My name is Bibekraj Tripathy.\n")
    file.write("I am a B.Tech CSE student.\n")
    file.write("I want to become a Full Stack Developer.\n")

print("Introduction written successfully.")
```

```python
#q2- Read and display file contents
with open("intro.txt", "r") as file:
    print(file.read())
```

```python
#q3- Append a new line 
with open("intro.txt", "a") as file:
    file.write("I love coding in Python.\n")

print("New line appended successfully.")
```

```python
with open("intro.txt", "r") as file:
    print(file.read())
```

```python
#q4 - Count numbers of  lines
with open("intro.txt", "r") as file:
    lines = file.readlines()
    print("Total lines:", len(lines))
```

```python
#q5 - Count number of words
with open("intro.txt", "r") as file:
    data = file.read()
    words = data.split()
    print("Total words:", len(words))
    
```

```python
#Q6 - Search a word
word = input("Enter word to search: ")

with open("intro.txt", "r") as file:
    data = file.read()

if word in data:
    print("found", word)
else:
    print("not found.", word)
```

```python
#Q7 -copy one file into another 
with open("intro.txt", "r") as source:
    data = source.read()

with open("copy.txt", "w") as destination:
    destination.write(data)

print("File copied successfully.")
```

```python
#Q8 - convert file content to uppercase
with open("intro.txt", "r") as file:
    data = file.read()

print(data.upper())
```

```python
#Q9 - store 5 student names
with open("students.txt", "w") as file:
    for i in range(5):
        name = input("Enter student name: ")
        file.write(name + "\n")

print("Student names stored successfully.")
```

```python
#Q10 - read alternate lines
with open("students.txt", "r") as file:
    lines = file.readlines()

for i in range(0, len(lines), 2):
    print(lines[i].strip())
```
