```python
import numpy as np

students = np.array(["Bibek", "Govind", "Hitesh", "Suraj", "Deba"])
subjects = np.array(["Math", "Science", "English", "Computer"])

marks = np.array([
    [85, 90, 78, 88],
    [72, 68, 75, 70],
    [95, 92, 90, 96],
    [60, 65, 58, 62],
    [88, 84, 91, 86]
])
print("Students:", students)
print("Subjects:", subjects)
print("Marks:\n", marks)
```

```python
average = np.mean(marks, axis=1)

print("Average Marks")
for name, avg in zip(students, average):
    print(name, ":", avg)
```

```python
highest = np.max(marks, axis=1)
lowest = np.min(marks, axis=1)

for i in range(len(students)):
    print(students[i],
          "Highest =", highest[i],
          "Lowest =", lowest[i])
```

```python
std = np.std(marks, axis=1)

for name, s in zip(students, std):
    print(name, ":", round(s,2))
```

```python
average = np.mean(marks, axis=1)

ranking = np.argsort(-average)

print("Ranking")
for rank, idx in enumerate(ranking, start=1):
    print(rank, students[idx], average[idx])
```

```python
student_avg = np.mean(marks, axis=1)

class_avg = np.mean(student_avg)

print("Class Average =", class_avg)
top_students = students[student_avg > class_avg]
print("Top Students:")
print(top_students)
```

```python
subject_average = np.mean(marks, axis=0)

for sub, avg in zip(subjects, subject_average):
    print(sub, ":", avg)
```
