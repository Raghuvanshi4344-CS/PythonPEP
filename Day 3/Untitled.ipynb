```python
#multiplication table
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
#Sum of First n Numbers
n = int(input("Enter n: "))

sum = 0

for i in range(1, n + 1):
    sum = sum + i

print("Sum =", sum)
```

```python
#factorial
num=int(input("Enter a number: "))

fact = 1
for i in range(1,num+1):
     fact = fact * i
print("Factorial = " , fact)

```
