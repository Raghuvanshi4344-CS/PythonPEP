```python
#count the number of vowels in a string
S = input("Enter a string: ")
count = 0
for ch in S.lower():
    if ch in "aeiou":
        count += 1

print("Number of vowels:", count)
```

```python
#check whether a string is a palindrome
s=input("Enter a string: ")
if s == s[::-1]:
    print("Palindrom")
else:
    print("Not Palindrom")
```

```python
#validate whether an email address has a format/domain
email = input("Enter email: ")

if "@" in email and "gmail" in email and email.endswith(".com"):
    print("Valid Email")
else:
    print("Invalid Email")
```

```python
#mask a phone number except last four digit 
phone = input("Enter phone number: ")

masked = "*" * (len(phone) - 4) + phone[-4:]

print("Masked Number:", masked)
```

```python
#find the second occurrence of a given word in a string
text = input("Enter a sentence: ")
word = input("Enter the word: ")

first = text.find(word)
second = text.find(word, first + 1)

if second != -1:
    print("Second occurrence at index:", second)
else:
    print("Second occurrence not found")
```

```python
#find the frequency of each character in a string 
s = input("Enter a string: ")

for ch in set(s):
    print(ch, ":", s.count(ch))
```

```python
#convert a snake_case string to Camelcase.
s = input("Enter snake_case string: ")

words = s.split("_")

camel = ""

for word in words:
    camel = camel + word.capitalize()

print("CamelCase:", camel)
```
