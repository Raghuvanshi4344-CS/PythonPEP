```python
import pandas as pd


data = {
    "Name": ["Bibek", "Hitesh", "Priyanshu", "Shiva", "Harsh"],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

# Head
print("Head")
print(df.head())

# Tail
print("\nTail")
print(df.tail())

# Sample
print("\nSample")
print(df.sample(2))

# Info
print("\nInfo")
df.info()

# Describe
print("\nDescribe")
print(df.describe())

# Shape
print("\nShape")
print(df.shape)

# Columns
print("\nColumns")
print(df.columns)

# Index
print("\nIndex")
print(df.index)

# Data Types
print("\nData Types")
print(df.dtypes)

# Value Counts
print("\nValue Counts (Age)")
print(df["Age"].value_counts())

# Unique Values
print("\nUnique Values (Age)")
print(df["Age"].unique())
```

```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [20, 21, 19, 22, 20],
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai"]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# -------------------- loc --------------------
print("\nloc - Select Row 2")
print(df.loc[2])

print("\nloc - Select Rows 1 to 3")
print(df.loc[1:3])

print("\nloc - Select Name and City Columns")
print(df.loc[:, ["Name", "City"]])

# -------------------- iloc --------------------
print("\niloc - Select First Row")
print(df.iloc[0])

print("\niloc - Select First 3 Rows")
print(df.iloc[0:3])

print("\niloc - Select Name and Age Columns")
print(df.iloc[:, 0:2])

# -------------------- isin() --------------------
print("\nisin() - Students from Delhi or Mumbai")
print(df[df["City"].isin(["Delhi", "Mumbai"])])

# -------------------- between() --------------------
print("\nbetween() - Age between 20 and 21")
print(df[df["Age"].between(20, 21)])

# -------------------- query() --------------------
print("\nquery() - Age > 20")
print(df.query("Age > 20"))

print("\nquery() - City == 'Mumbai'")
print(df.query("City == 'Mumbai'"))

print("\nquery() - Age >= 20 and City == 'Delhi'")
print(df.query("Age >= 20 and City == 'Delhi'"))
```

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", None, "David"],
    "Age": [20, None, 19, 22],
    "City": ["Delhi", "Mumbai", None, "Chennai"]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# dropna()
print("\nAfter dropna()")
print(df.dropna())

# fillna()
print("\nAfter fillna()")
print(df.fillna("Unknown"))
```

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Bob", "David"],
    "Age": [20, 21, 21, 22],
    "City": ["Delhi", "Mumbai", "Mumbai", "Chennai"]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# replace()
print("\nreplace()")
print(df.replace("Delhi", "Bhubaneswar"))

# rename()
print("\nrename()")
print(df.rename(columns={"Name": "Student_Name", "Age": "Student_Age"}))

# duplicated()
print("\nduplicated()")
print(df.duplicated())

# drop_duplicates()
print("\ndrop_duplicates()")
print(df.drop_duplicates())
```
