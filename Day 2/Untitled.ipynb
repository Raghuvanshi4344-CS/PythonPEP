```python
#firsttime user -FIRST100 where user will get 100 off on order above 399

# user with order more than  total 5000 rupeess in his account   = get 199 off on next order if the order amount is more than 1000 
#user with order more than  total 10000 rupeess in his account   = get 399 off on next order if the order amount is more than 2500 
# user with order more than  total 20000 rupeess in his account   = get 499 off on next order if the order amount is more than 5000 
# if the user is using his account after 6 month - welcomeback = where user will get a discount of 149 on order above than 449


total = int(input("Enter total purchase amount: "))
order = int(input("Enter current order amount: "))
first = input("First time user (yes/no): ")
back = input("Using after 6 months (yes/no): ")
if first == "yes" and order > 399:
    print("Discount = 100")

elif back == "yes" and order > 449:
    print("Discount = 149")

elif total > 20000 and order > 5000:
    print("Discount = 499")

elif total > 10000 and order > 2500:
    print("Discount = 399")

elif total > 5000 and order > 1000:
    print("Discount = 199")

else:
    print("No Discount")
```

```python
#build a system like atm machine
balance = 5000
pin=int(input("Enter the pin: "))

if pin == 1234:

    print("1. Check Balance")
    print("2. Withdraw")

    choice = int(input("Enter choice: "))

    if choice == 1:
        print("Balance =", balance)

    elif choice == 2:
        amount = int(input("Enter amount: "))

        if amount <= balance:
            print("Cash Withdrawn")
            print("Remaining Balance =",balance - amount)

        else:
            print("Insufficient Balance")

    else:
        print("Invalid Choice")

else:
    print("Wrong PIN")



```
